# 1.Importind all Dependencies

```python
import os
import cv2
import tensorflow as tf
import json
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, Reshape, Dropout
from tensorflow.keras.applications import ResNet152V2
from tensorflow.keras.models import load_model
```

# 2. Converting Data Into Datasets

```python
def load_images(x):
    byte_img=tf.io.read_file(x)
    img=tf.io.decode_jpeg(byte_img)
    return img
```

Training Dataset is used to train the model

```python
train_images=tf.data.Dataset.list_files("Data2\\Train\\Images\\*.jpg",shuffle=False)
train_images=train_images.map(load_images)
train_images=train_images.map(lambda x: tf.image.resize(x,(250,250)))
train_images=train_images.map(lambda x:x/255)
```

Validation Dataset is used for validation process during the training process

```python
val_images=tf.data.Dataset.list_files("Data2\\Val\\Images\\*.jpg",shuffle=False)
val_images=val_images.map(load_images)
val_images=val_images.map(lambda x: tf.image.resize(x,(250,250)))
val_images=val_images.map(lambda x:x/255)
```

```python
def load_labels(label_path):
    with open(label_path.numpy(),"r",encoding='utf-8')as f:
        label=json.load(f)
    return [label['keypoints']]
```

```python
train_labels = tf.data.Dataset.list_files('Data2\\Train\\Labels\\*.json', shuffle=False)
train_labels = train_labels.map(lambda x: tf.py_function(load_labels, [x], [tf.float16]))
```

```python
train_labels.as_numpy_iterator().next()
```

```python
val_labels=tf.data.Dataset.list_files("Data2\\Val\\Labels\\*.json",shuffle=False)
val_labels=val_labels.map(lambda x: tf.py_function(load_labels,[x],[tf.float16]))
```

```python
val_labels.as_numpy_iterator().next()
```

Above we can see that we have taken Images and Labels separately. In the below step, we are zipping the Images and Labels together.

```python
train = tf.data.Dataset.zip((train_images, train_labels))
train = train.shuffle(8400)
train = train.batch(16)
train = train.prefetch(4)
```

```python
val=tf.data.Dataset.zip((val_images,val_labels))
val=val.shuffle(1800)
val=val.batch(16)
val=val.prefetch(4)
```

```python
data_sample=train.as_numpy_iterator()
```

```python
res=data_sample.next()
```

```python
res[1][0]
```

# 3. Visualizing Data

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx in range(4): 
    sample_image = res[0][idx]
    sample_coords = res[1][0][idx]
    
    cv2.circle(sample_image, tuple(np.multiply(sample_coords[:2], [250,250]).astype(int)), 2, (255,0,0), -1)
    cv2.circle(sample_image, tuple(np.multiply(sample_coords[2:], [250,250]).astype(int)), 2, (0,255,0), -1)
    
    ax[idx].imshow(sample_image)
```

# 4. Declaring Model

```python
model = Sequential([
    Input(shape=(250,250,3)), 
    ResNet152V2(include_top=False, input_shape=(250,250,3)),
    Conv2D(512, 3, padding='same', activation='relu'),
    Conv2D(512, 3, padding='same', activation='relu'),
    Conv2D(256, 3, 2, padding='same', activation='relu'),
    Conv2D(256, 2, 2, activation='relu'),
    Dropout(0.05),
    Conv2D(4, 2, 2),
    Reshape((4,))
])
```

```python
model.summary()
```

```python
optimiser=tf.keras.optimizers.Adam(learning_rate=0.001,decay=0.0007)
loss=tf.keras.losses.MeanSquaredError()
```

```python
model.compile(optimiser,loss)
```

# 5. Training the Model

```python
hist=model.fit(train,epochs=25,validation_data=val)
```

From the above Training, we can see that the loss and Val_Loss is steaply decreasing, which is a sign that the model is training well. But the model is not completely ready for real time use, because the data given was not that clearly captured. But we can implement the same process on clear and large dataset for a real time model.

```python
model.save("model.h5")
```

```python
model = load_model('model.h5')
```

# 6. Testing the model in real time

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    _ , frame = cap.read()
    
    frame = frame[50:500,50:500,:] 
    rgb_img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(rgb_img, (250,250))
    
    yhat = model.predict(np.expand_dims(resized/255,0))
    sample_coords = yhat[0,:4]
    
    cv2.circle(frame, tuple(np.multiply(sample_coords[:2], [450,450]).astype(int)), 2, (255,0,0), -1)
    cv2.circle(frame, tuple(np.multiply(sample_coords[2:], [450,450]).astype(int)), 2, (0,255,0), -1)
    
    cv2.imshow('EyeTrack', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
